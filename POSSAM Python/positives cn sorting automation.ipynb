```python
#this is where barcodes are scanned in Alinity-mapping style(A1-A12, B1-B12, ..., H1-H12)

list_of_barcodes = '''




'''

list_of_barcodes = list_of_barcodes.strip().split('\n')
```

```python
THRESHOLD_MONROVIA = 26
# THRESHOLD_EMPLOYEE = 26
THRESHOLD_NEWMEXICO = 26
THRESHOLD_APH = 26
THRESHOLD_HOUSTON = 26

SORTING_APH = True
SORTING_HOUSTON = False
SORTING_NEWMEXICO = False
```

```python
aph_zipcodes = list(map(str, [76574, 78605, 78610, 78612, 78613, 78615, 78616, 78617, 78620, 78621, 78626, 78628, 78634, 78640, 78641, 
                78642, 78644, 78645, 78652, 78653, 78654, 78660, 78663, 78664, 78665, 78669, 78681, 78701, 78702, 78703, 
                78704, 78705, 78712, 78717, 78719, 78721, 78722, 78723, 78724, 78725, 78726, 78727, 78728, 78729, 78730, 
                78731, 78732, 78733, 78734, 78735, 78736, 78737, 78738, 78739, 78741, 78742, 78744, 78745, 78746, 78747, 
                78748, 78749, 78750, 78751, 78752, 78753, 78754, 78756, 78757, 78758, 78759]))
houston_testing_site_ids = [31030, 30437, 17954, 29966, 28551, 6966, 6967, 8861, 8868, 29344, 24534, 29317, 22753, 29343, 
                            29266, 10818, 27233, 31580, 31581, 26823, 31579, 24459, 28703, 28862, 25566, 30673, 31172, 
                            15333, 29267, 10040, 24703, 28748, 24544, 31614, 17836, 28775, 28774, 24541, 30596, 27585]

well_pos = [i+str(j) for i in 'ABCDEFGH' for j in range(1,13)][:len(list_of_barcodes)]
samples = pd.DataFrame({'barcodes': list_of_barcodes, 'well_pos': well_pos, 'sorting_order': range(len(list_of_barcodes))})

with os.scandir() as it:
    possible_ct_files, possible_phi_files, possible_ts_files = list(), list(), list()
    for entry in it:
        if entry.name.startswith('query_result'):
            possible_ct_files.append(entry.name)
        elif entry.name.startswith('test_kits_possam_doh_phi__') and SORTING_APH:
            possible_phi_files.append(entry.name)
ct_csv = max(possible_ct_files, key = os.path.getmtime)
if SORTING_APH:
    phi_csv = max(possible_phi_files, key = os.path.getmtime)

if SORTING_APH:
    phi = pd.read_csv(phi_csv)
    phi.fillna(value={'patients_zipcode':10000}, inplace = True)
    phi.loc[:, 'patients_zipcode'] = phi.patients_zipcode.apply(lambda x: str(x)[:5])
ct = pd.read_csv(ct_csv)
ct.loc[:, 'Completed At'] = ct['Completed At'].apply(pd.to_datetime)
ct.loc[:, 'Alinity Results → Created At'] = ct['Alinity Results → Created At'].apply(pd.to_datetime)
#getting the most recent results only
ct = ct.sort_values('Alinity Results → Created At').drop_duplicates('Test Kits → Barcode')
ct.fillna(value = {'Result':'Missed Exception'}, inplace = True)

if SORTING_APH:
    df = samples.set_index('barcodes')\
    .join(ct[['Test Kits → Barcode', 'Alinity Results → Cn', 'Result', 'Testing Sites → State', 'Facility ID']]\
         .set_index('Test Kits → Barcode'))\
    .join(phi[['barcode_id', 'patients_zipcode']]\
          .set_index('barcode_id'))\
    .reset_index()
else:
    df = samples.set_index('barcodes')\
    .join(ct[['Test Kits → Barcode', 'Alinity Results → Cn', 'Result', 'Testing Sites → State', 'Facility ID']]\
         .set_index('Test Kits → Barcode'))\
    .reset_index()

def sorting(row):
    if row['Result'] == 'POSITIVE':
        #Facility ID = 'Curative Employee Testing'
        if row['Facility ID'] == 10 and row['Alinity Results → Cn'] > 0: 
            return 'Employee'
        elif SORTING_NEWMEXICO and\
        row['Testing Sites → State'] == 'NM' and\
        row['Alinity Results → Cn'] < THRESHOLD_NEWMEXICO and\
        row['Alinity Results → Cn'] > 0:
            return 'Strong New Mexico'
        elif SORTING_HOUSTON and\
        row['Testing Site ID'] in houston_testing_site_ids and\
        row['Alinity Results → Cn'] < THRESHOLD_HOUSTON and\
        row['Alinity Results → Cn'] > 0:
            return 'Houston'
        elif SORTING_APH\
        and row['patients_zipcode'] in aph_zipcodes\
        and row['Alinity Results → Cn'] < THRESHOLD_APH\
        and row['Alinity Results → Cn'] > 0:
            return 'Austin Public Health'
        elif row['Alinity Results → Cn'] < THRESHOLD_MONROVIA\
        and row['Alinity Results → Cn'] > 0:
            return 'Strong'
        elif row['Alinity Results → Cn'] >= THRESHOLD_MONROVIA:
            return 'Weak'
        else:
            return 'error1'
    elif row['Result'] in ['NEGATIVE', 'INDETERMINATE']:
        return 'Negative rack'
    elif row['Result'] in ['QNS', 'TNP', 'SNR']:
        return 'QNS/TNP'
    elif row['Result'] == 'Missed Exception':
        return 'Missed Exception'
    return 'error2'

df['sorting_rack'] = df.apply(lambda row: sorting(row), axis = 1)

df2 = df.sort_values('sorting_order')[['barcodes', 'well_pos', 'sorting_rack', 'Alinity Results → Cn', 'Result', 
                                       'Testing Sites → State', 'Facility ID']].copy()
```

```python
# #DEMO PURPOSE


df2 = pd.DataFrame({'sorting_rack': ['Employee',
                                     'Strong New Mexico', 
                                     'Houston',
                                     'Austin Public Health', 
                                     'Strong', 
                                     'Weak', 
                                     'Negative rack',
                                     'QNS/TNP',
                                     'Missed Exception',
                                     'error1',
                                     'error2']})

print(df2.sorting_rack.value_counts())

def color_scheme(sr):
    return {'Employee': 'red',
           'Strong New Mexico': 'orangered',
#            'Weak New Mexico': 'gold',
           'Strong': 'royalblue',
           'Weak': 'lightskyblue',
           'Houston': 'mediumpurple',
           'Austin Public Health': 'forestgreen',
           'Negative rack': 'silver',
           'QNS/TNP': 'dimgrey',
           'error1': 'white',
           'error2': 'white',
           'Missed Exception': 'fuchsia'}[sr]

def acronym(sr):
    return {'Employee': 'EMP',
           'Strong New Mexico': 'Str\nNM',
#            'Weak New Mexico': 'Wk\nNM',
           'Strong': 'Str',
           'Weak': 'Wk',
           'Houston': 'Hou',
           'Austin Public Health': 'APH',
           'Negative rack': 'Neg',
           'QNS/TNP': 'QNS/\nTNP',
           'error1': 'Err1',
           'error2': 'Err2',
           'Missed Exception': 'ME'}[sr]


radius = 2
spacing = 1

df3 = pd.DataFrame({'sorting_rack': df2.copy().sorting_rack,
                   'grid_xpos': (list(range(12))*8)[:len(df2)],
                   'grid_ypos': [i for i in range(7, -1, -1) for _ in range(12)][:len(df2)]})
df3.loc[:, 'plt_xpos'] = df3.grid_xpos.apply(lambda x: (2*radius+spacing)*x)
df3.loc[:, 'plt_ypos'] = df3.grid_ypos.apply(lambda y: (2*radius+spacing)*y)
df3.loc[:, 'well_color'] = df3.sorting_rack.apply(color_scheme)
df3.loc[:, 'sorting_rack_acronym'] = df3.sorting_rack.apply(acronym)

fig, ax1 = plt.subplots(figsize = (30, 20), facecolor = 'white')

#assuming bottom left is 0,0

ax1.set_xlim(left = -2*radius-spacing, right = 23*radius + 12*spacing)
ax1.set_ylim(bottom = -2*radius-spacing, top = 15*radius + 8*spacing)

for i in range(len(df3)):
    ax1.add_patch(plt.Circle(xy = (df3.loc[i].plt_xpos, df3.loc[i].plt_ypos),
                            radius = radius,
                            edgecolor = 'black', 
                            facecolor = df3.loc[i].well_color))
    plt.text(x = df3.loc[i].plt_xpos, y = df3.loc[i].plt_ypos,
            s = df3.loc[i].sorting_rack_acronym,
            ha = 'center', va = 'center', fontsize = 30)
for x,y in map(lambda x: ((2*radius+spacing)*x[0], (2*radius+spacing)*x[1]), 
               [(x,y) for y in range(7, -1, -1) for x in range(12)][len(df3):]):
    ax1.add_patch(plt.Circle(xy = (x,y),
                            radius = radius,
                            edgecolor = 'black',
                            facecolor = 'white'))    

for y, txt in list(zip(map(lambda x: (2*radius+spacing)*x, range(7, -1, -1)), 'ABCDEFGH')):
    plt.text(x = -2*radius, y = y,
           s = txt, ha = 'center', va = 'center', fontsize = 30)

for x, txt in list(zip(map(lambda x: (2*radius+spacing)*x, range(12)), map(str, range(1,13)))):
    plt.text(x = x, y = -2*radius,
            s = txt, ha = 'center', va = 'center', fontsize = 30)

plt.vlines(x = [-radius-0.5*spacing+x*(spacing+2*radius) for x in range(13)], 
           ymin = -2*radius-0.5*spacing, ymax = 15*radius+7.5*spacing, colors = 'black', linewidths = 2)

plt.hlines(y = [-radius-0.5*spacing+y*(spacing+2*radius) for y in range(9)],
          xmin = -2*radius-0.5*spacing, xmax = 23*radius+11.5*spacing, colors = 'black', linewidths = 2)
 
plt.axis('off')
plt.show()

df2
```
